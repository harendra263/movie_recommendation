```python
movies_df= pd.read_csv('tmdb_5000_movies.csv')
credit_df = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie_col_renamed= movies_df.rename(index=str , columns={'id':'movie_id'})
```

```python
movie_df_merge=  movie_col_renamed.merge(credit_df, on ='movie_id')
movie_df_merge.head()
```

```python
movie_df_merge.columns
```

```python
movie_df_cleaned=movie_df_merge.drop(columns=['homepage','status','production_countries','title_x','title_y'])
movie_df_cleaned.head()
```

```python
# Calculate the components based on above formula
R = movie_df_cleaned['vote_average']
v = movie_df_cleaned['vote_count']
C = movie_df_cleaned['vote_average'].mean()
m = movie_df_cleaned['vote_count'].quantile(0.70)
```

```python
movie_df_cleaned['weighted_average']=((R*v) + (C*m))/(v+m)
```

```python
movie_df_cleaned.head()
```

```python
movie_sorted_ranking = movie_df_cleaned.sort_values('weighted_average' , ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
movie_sorted_ranking['weighted_average'].min(),movie_sorted_ranking['weighted_average'].max()
```

```python
weight_average = movie_sorted_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x = weight_average['weighted_average'].head(10) , y = weight_average['original_title'].head(10) , data=weight_average)
plt.xlim(4,10)
plt.title("Best movies by average votes", weight ='bold')
plt.xlabel('Weighted Average Score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.savefig('Best Movies.png')
```

```python
popularity = movie_sorted_ranking.sort_values('popularity', ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x= popularity['popularity'].head(10),y=popularity['original_title'].head(10), data=popularity)
plt.title('Most popular Movies', weight='bold')
plt.xlabel('popularity', weight=10)
plt.ylabel('Movie_title',weight=10)
plt.savefig('Popular Movie.png')
```

### Recommendation based on scaled weighted average and popularity score (Priority is given to both)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movie_df_cleaned[['weighted_average','popularity']])
movie_normalized_df = pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
# movie_df_cleaned.drop(['normalized_weight_average','normalized_popularity'], axis=1, inplace=True)|
```

```python
movie_df_cleaned[['normalized_weighted_average','normalized_popularity']]=movie_normalized_df
movie_df_cleaned.head()
```

```python
movie_df_cleaned['score']=movie_df_cleaned['normalized_weighted_average']*0.5 + movie_df_cleaned['normalized_popularity']*0.5
movie_scored_df = movie_df_cleaned.sort_values('score', ascending=False)
movie_scored_df[['original_title','normalized_weighted_average','normalized_popularity', 'score']].head(10)
```

```python
score_df = movie_df_cleaned.sort_values('score', ascending=False)

plt.figure(figsize=(20,6))

sns.barplot(x = score_df['score'].head(10),y= score_df['original_title'].head(10), data=score_df)
plt.title('Most Rated and popular Blend Movies')
plt.xlabel('score', weight='bold')
plt.ylabel('movies',weight='bold')
plt.savefig('Most Rated and popular Movie.png')

```
